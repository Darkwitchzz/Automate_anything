```python
import yfinance as yf

ticker_symbol = "BTC-USD"

ticker = yf.Ticker(ticker_symbol)

start = False
end = None
# start="2020-01-01"
# end="2023-01-01"

if start:
  historical_data = ticker.history(interval="1d", start=start, end=end)

else:
  historical_data = ticker.history(period="max", interval="1d")

historical_data.drop(columns=["Dividends", "Stock Splits"], inplace=True)

historical_data.to_csv(f"{ticker_symbol}.csv")

print(historical_data)


```
